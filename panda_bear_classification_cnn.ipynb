```python

import os

zip_path = "/content/archive (1).zip"
extract_path = "/content/panda_bear_classification"
```

```python
for root, dirs, files in os.walk(extract_path):
    print(root)
    for d in dirs:
        print("  DIR:", d)
    for f in files:
        print("  FILE:", f)
```

```python
train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

val_datagen = ImageDataGenerator(rescale=1./255)

# Train set
train_set = train_datagen.flow_from_directory(
    '/content/panda_bear_classification/PandasBears/Train',
    target_size=(128, 128),
    batch_size=32,
    class_mode='binary'
)


val_set = val_datagen.flow_from_directory(
    '/content/panda_bear_classification/PandasBears/Test',
    target_size=(128, 128),
    batch_size=32,
    class_mode='binary'
)
```

```python
model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPooling2D(2,2),

    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),


    Flatten(),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])


model.fit(
    train_set,
    validation_data=val_set,
    epochs=10
)
```

```python
import numpy as np
from tensorflow.keras.preprocessing import image

img_path = '/content/panda_bear_classification/PandasBears/Test/Pandas/266.jpeg'
img = image.load_img(img_path, target_size=(128, 128))
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.0

prediction = model.predict(img_array)

if prediction[0][0] > 0.5:
    print("Predicted: Bear 🐻")
else:
    print("Predicted: Panda 🐼")
```
